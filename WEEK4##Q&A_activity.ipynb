# TO IMPORT THE LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# TO READ THE EXCEL FILE

```python
df=pd.read_excel('Telco-Customer-Churn.xlsx')
df.head()
```

# DETAILS ABOUT THE GIVEN DATASET

```python
df.info()
```

# 1.COMPARING CHURN COUNT WITH RESPECT TO GENDER

```python
plt.figure(figsize=(15,15))
sns.countplot(x='Churn',data=df,hue='gender')
plt.title('COMPARING CHURN WITH RESPECT TO GENDER',fontsize=20)
plt.show()
```

# 2. VISUALIZATION ABOUT FEMALE SENIOR CITIZENS IN THE GIVEN DATASET

```python
df1=df[(df['SeniorCitizen']==1)&(df['gender']=='Female')]
df1
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x=df1['SeniorCitizen'])
plt.title('TOTAL NUMBER OF FEMALE SENIOR CITIZENS',fontsize=15)
plt.xlabel('FEMALE SENIOR CITIZEN',fontsize=15)
```

```python
df1.groupby(['SeniorCitizen','gender'])['gender'].count()
```

```python
##from the above graph we can understand that there was a total of 568 senior women citizens in the given data set
```

# 3. COMPARING TENURE WITH TOTAL CHARGES

```python
plt.figure(figsize=(15,15))
sns.scatterplot(x='tenure',y='TotalCharges',data=df,hue='SeniorCitizen')
plt.title('COMPARING TENURE WITH RESPECT TO TOTAL CHARGES',fontsize=20)
plt.show()
```

```python
##from the plot given above we can come to the conclusion that as the tenure increases along with the increase in total charges
```

# 4. CONTRACTS PREFERED BY SENIOR CITIZEN

```python
plt.figure(figsize=(15,15))
sns.countplot(x=df[df['SeniorCitizen']==1]['Contract'],)
plt.title('CONTRACTS PREFFERED BY SENIOR CITIZENS',fontsize=15)
```

```python
##from the graphical representation of the given data we can conclude that the month-to-month contract was most preffered by the customers and the two year contract was the least preffered
```

# 5. PLOT OF THE PAYMENT METHODS USED

```python
plt.figure(figsize=(15,15))
sns.countplot(x='PaymentMethod',data=df,)
plt.title(' PLOT OF THE PAYMENT METHODS USED',fontsize=15)
```

```python
##from the visualization about the given dataset ,we can conclude that electronic check was the most used payment method and credit card was the least used 
```
